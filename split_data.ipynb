## Split data to tran, validation, test

```python
#Import library 

import os
import glob
import random
import math
import sh
import shutil
```

```python
print(os.getcwd())
```

```python
# Weight of split data
TRAIN_WEIGHT = 0.65
TEST_WEIGHT = 0.2
VALIDATION_WEIGHT = 0.15

# Define the root of data
HOME = os.getcwd()

# Direction of split data
DIR_NAME = 'split_data'
```

```python
# Gather and randomlize subjects

subjects = []
for img_dir in glob.glob('*'):
  if os.path.isdir(img_dir):
    subjects.append(img_dir)

random.shuffle(subjects)
```

```python
print(len(subjects))
```

```python
# Split subjects
size = len(subjects)

train = subjects[0: math.floor(TRAIN_WEIGHT*size)]
validation = subjects[math.floor(TRAIN_WEIGHT*size) : math.floor((TRAIN_WEIGHT + VALIDATION_WEIGHT)*size)]
test = subjects[math.floor((TRAIN_WEIGHT + VALIDATION_WEIGHT)*size) : ]
```

```python
print(train)
print(validation)
print(test)
```

```python
# Check the overlap
x = train + validation + test
x = set(x)

assert(len(x) == len(subjects))
```

```python
# Create train, validation, test folder
os.chdir('..')
if os.path.exists(DIR_NAME):
  shutil.rmtree(DIR_NAME)
os.mkdir(DIR_NAME)
os.chdir(DIR_NAME)

ln = sh.Command('ln')

# Train folder
os.mkdir('train')
os.chdir('train')
for subject in train:
  folder_path = f'{HOME}/{subject}'
  ln('-sfv', folder_path, subject)
os.chdir('..')

# Validation folder
os.mkdir('validation')
os.chdir('validation')
for subject in validation:
  folder_path = f'{HOME}/{subject}'
  ln('-sfv', folder_path, subject)
os.chdir('..')

# Test folder
os.mkdir('test')
os.chdir('test')
for subject in test:
  folder_path = f'{HOME}/{subject}'
  ln('-sfv', folder_path, subject)
os.chdir('..')


```
